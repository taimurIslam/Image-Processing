```python
%matplotlib inline
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
# The image has been correctly loaded by openCV as a numpy array,  
# but the color of each pixel has been sorted as BGR (Blue, Green, Red).
image = cv2.imread('image.jpg')
# cvtColor is the openCV function which changes the color space of an image
original_color_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(original_color_image)
plt.show()
```
