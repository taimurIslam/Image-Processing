```python
%matplotlib inline
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

# Read and Show Image

```python
# The image has been correctly loaded by openCV as a numpy array,  
# but the color of each pixel has been sorted as BGR (Blue, Green, Red).
image = cv2.imread('image.jpg')
# cvtColor is the openCV function which changes the color space of an image
original_color_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(original_color_image)
plt.show()
```

### Access the pixel values of a particular point (row, column)

```python
# Shape of the image
print(original_color_image.shape)
#access pixel values of a particular position
px = original_color_image[500,400]
px
```

# Change the pixel values of a particular point (row, column)

```python
original_color_image[500,400] = [255, 255, 255]
print(original_color_image[500,400])
```

```python
# 0=R
# 1=G
# 2=B
original_color_image.itemset((500,400,0),254) 
print(original_color_image[500,400])
```

```python
import cv2
import numpy as np

cap = cv2.VideoCapture(0)

while(1):

    # Take each frame
    _, frame = cap.read()

    # Convert BGR to HSV
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)

    # define range of blue color in HSV
    lower_blue = np.array([110,50,50])
    upper_blue = np.array([130,255,255])

    # Threshold the HSV image to get only blue colors
    mask = cv2.inRange(hsv, lower_blue, upper_blue)

    # Bitwise-AND mask and original image
    res = cv2.bitwise_and(frame,frame, mask= mask)

    cv2.imshow('frame',frame)
    cv2.imshow('mask',mask)
    cv2.imshow('res',res)
    k = cv2.waitKey(5) & 0xFF
    if k == 27:
        break

cv2.destroyAllWindows()
```
